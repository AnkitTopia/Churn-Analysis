```python
#1) Credit Score
```

```python
sns.kdeplot(data=df, x='credit_score', hue='churn', fill=True, palette={0: 'green', 1: 'red'})

plt.show()
```

```python
#FROM THE ABOVE GRAPH WE CAN SEE THAT THE CREDIT SCORE FOR BOTH CHURN AND NOT CHURN CUSTOMERS ARE ALMOST SAME THAT INDICATES THAT
# CREDIT SCORE IS NOT CONTRIBUTING AT ALL IN DETERMINIG WHETHER THE CUSTOMER WILL LEAVE OR STAY.
```

```python
df.drop(['credit_score'],axis=1,inplace=True) #deleting the credit_score column as it is not contributing to our analysis.
```

```python
#2) Country
```

```python
ax=sns.countplot(data=df, x= "country",hue='churn')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# from above chart we can see that Customers in Germany are leaving at exactly double the rate of customers in France and Spain.
```
